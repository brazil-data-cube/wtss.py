```python
coverage = service['MOD13Q1-6']
coverage
```

```python
timeseries = coverage.ts(attributes = ['NDVI','EVI'],
                        geom = {"type":"Polygon","coordinates":[[[-54,-12],[-53.99,-12],[-53.99,-11.99],[-54,-11.99],[-54,-12]]]},
                        start_datetime = '2017-01-01',
                        end_datetime = '2017-02-01')

timeseries
```

```python
timeseries.NDVI
```

```python
summarize = coverage.summarize( attributes = ['NDVI','EVI'],
                                geom = {"type":"Polygon","coordinates":[[[-54,-12],[-53.99,-12],[-53.99,-11.99],[-54,-11.99],[-54,-12]]]},
                                start_datetime = '2017-01-01', 
                                end_datetime = '2018-03-01', 
                                applyAttributeScale = False)

summarize
```

```python
summarize.timeline
```

```python
summarize.NDVI.mean
```

```python
# The user can set the list of desired attributes to plot and the aggregation metric
summarize.plot(
    attributes=['NDVI','EVI'],
    aggregation='mean'
)
```

```python
# The other plot feature is the mean+std and mean-std plot, where the user need to set only the attribute string
summarize.plot_mean_std(
    attribute='NDVI'
)
```

```python
aggregation_dataframe = summarize.pandas_dataframe()
aggregation_dataframe
```

```python
# Export dataframe to csv
aggregation_dataframe.to_csv('aggregation_dataframe.csv')
```
